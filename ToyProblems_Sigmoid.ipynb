```python
import data_problems_examples
from neural_network_recreate_01 import *
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
type_chosen = 'xor'

X_train, Y_train = data_problems_examples.generate_problem(type=type_chosen)
Y_train = Y_train.reshape(Y_train.shape[0], 1)

X_test, Y_test = data_problems_examples.generate_problem(type=type_chosen)
```

```python
plt.figure(figsize=(6, 6))
plt.scatter(X_train[:, 0], X_train[:, 1], c=Y_train, s=30)
plt.grid()
```

```python
nn = NeuralNetwork(layers=[
    SigmoidLayer(3, bias=False),
    SigmoidLayer(1, bias=False)
], input_bias=True, loss_function='MSE', regular_type='l2', alpha=1e-4)
```

```python
nn.fit(X_train, Y_train, n_epoch=200, batch_size=15, learning_params=(25, 0.75, 65))
plt.figure(figsize=(7, 5))
plt.plot(nn.error)
plt.xlabel('epoch')
plt.ylabel(nn.loss_function)
plt.grid()
```

```python
Y_predicted = nn.predict(X_test, batch_size=None)
Y_predicted = Y_predicted.reshape(Y_predicted.shape[0])
```

```python
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True, figsize=(13, 6))
ax1.scatter(X_test[:, 0], X_test[:, 1], c=Y_predicted > 0.5, s=30)
ax1.grid()
ax2.scatter(X_test[:, 0], X_test[:, 1], c=Y_predicted, s=30)
ax2.grid()
```
